```python
# Write a program to perform silly sort for given array

# Silly sort : sort the array left half and right half separatly.


# Constraints:
# Input :                 First line of input consits of one integer value reprsents array size.

#                              From Second line it consists of few integer values .

# Output :              Print the array after perfoming silly sort 

# Constraints :       If the given array size is odd then sort left half and right half separatly and don't disturb the middle element


# Example:
# Input1 :             6

# #                           9

# #                           10

# #                            7

# #                            13

# #                             0 

# #                             1 

# # Output1 :          7 9 10 0 1 13

# # Input2 :             9

# #                            10

# #                             8

# #                            13

# #                             4

# #                             100

# #                             18

# #                              4 

# #                              0

# #                              1

# # Output2 :           4 8 10 13 100 0 1 4 18 

# def sort(arr):
#     mid=len(arr)//2
#     left=arr[:mid]
#     right=arr[mid:]
#     a1=[]
#     a2=[]
#     max1=left[0]
#     max2=right[0]
#     for i in range(len(left)):
        
#         if left[i]>max1:
#             max1=left[i]
#         a1.append(max1)
#     for i in range(len(right)):
        
#         if right[i]>max2:
#             max2=right[i]
#         a2.append(max2)
#     return a1,a2
    

# arr=[9,10,7,13,0,1]
# print(sort(arr))

def bubble(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr

def lrsort(arr):
    mid=len(arr)//2
    if len(arr)%2==0:
        left=bubble(arr[:mid])
        right=bubble(arr[mid:])

        result=left+right
        return " ".join(map(str,result))
    else:
        left=bubble(arr[:mid])
        middle=[arr[mid]]
        right=bubble(arr[mid+1:])
        result=left+middle+right
        return " ".join(map(str,result))

arr=[9,10,7,13,0,1,2]
print(lrsort(arr))


```
