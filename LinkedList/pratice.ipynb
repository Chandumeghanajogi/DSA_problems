```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Linked:
    # Build linked list in reverse order (so [4,5] becomes 5->4)
    def build(self, l1):
        self.head = None
        for i in l1:
            new_node = node(i)
            new_node.next = self.head
            self.head = new_node
        return self.head

    # Add two numbers represented by linked lists
    def addnum(self, l1, l2):
        dummy=curr = node(0)
        carry = 0
        while l1 or l2 or carry:
            sum_val = carry
            if l1:
                sum_val += l1.data
                l1 = l1.next
            if l2:
                sum_val += l2.data
                l2 = l2.next
            curr.next = node(sum_val % 10)
            curr = curr.next
            carry = sum_val // 10
        return dummy.next

    # Display linked list
    def display(self, res):
        temp = res
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

# Example test
l1 = [4, 5]      # represents 45
l2 = [3, 4, 5]   # represents 345

LL = Linked()
list1 = LL.build(l1)
list2 = LL.build(l2)
res = LL.addnum(list1, list2)

print("Output linked list:")
LL.display(res)   # Expected: 0 9 3
```
